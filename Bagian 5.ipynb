# Seleksi weekdays & weekends pada Data Deret Waktu

```python
weekdays_df = df[df.index.dayofweek.isin([0, 1, 2, 3, 4])]
weekdays_df.head(7)
```

```python
weekends_df = df[df.index.dayofweek.isin([5,6])]
weekends_df.head(7)
```

# Cummulative count pada Data Frame

```python
d = {'penjual':['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'],
     'barang':['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor', 'laptop', 'monitor']}
    
df = pd.DataFrame(d)
df
```

```python
df['count_tiap_penjual'] = df.groupby('penjual').cumcount() + 1
df
```

```python
df['count_tiap_barang'] = df.groupby('barang').cumcount() + 1
df
```

```python
df['count_pasangan_kolom'] = df.groupby(['penjual', 'barang']).cumcount() + 1
df
```

# F-string & For Loop untuk membuat kolom baru

```python
d = {'penjual':['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'],
     'barang':['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor', 'laptop', 'monitor']}
    
df = pd.DataFrame(d)
df
```

```python
cols = ['penjual', 'barang']

for col in cols :
    df[f'count_tiap_{col}'] = df.groupby(col).cumcount() + 1
    
df
```

# Membuat kategori baru berdasarkan threshold

```python
d = {'hobi':['jogging', 'mancing', 'renang',
             'mancing', 'mancing', 'baca',
             'baca', 'mancing', 'fotografi',
             'mancing', 'camping']}

df = pd.DataFrame(d)
df
```

```python
df['hobi'].value_counts()
```

```python
persentase = df['hobi'].value_counts(normalize=True)
persentase
```

```python
threshold = 0.1 
hobi_lain = persentase[persentase < threshold].index
hobi_lain
```

```python
df['hobi'] = df['hobi'].replace(hobi_lain, 'Lainnya')
df['hobi']
```

```python
df['hobi'].value_counts(normalize=True)
```
