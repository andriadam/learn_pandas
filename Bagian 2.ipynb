# Konversi nilai numerik ke dalam kategori

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)),columns=cols)

df
```

# Menggabungkan (merge) 2 Pandas data frame

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1,4])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])
df2
```

```python
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```

# Memecah nilai string menjadi beberapa kolom

```python
data = {'nama':['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'],
        'tempat_kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}
df = pd.DataFrame(data)
df
```

```python
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

```python
df[['kota', 'provinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```

# Resampling pada time series data

```python
n_rows = 365 * 24
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df
```

```python
df.resample('M')['col1'].sum().to_frame()
```

```python
df.resample('D')['col1'].sum().to_frame()
```

# Formatting tampilan Data Frame

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
df['omset'] = df['omset'] * 100_000
df['operasional'] = df['operasional'] * 10_000
df
```

```python
df.index = pd.util.testing.makeDateIndex(n_rows, freq='D')
df = df.reset_index()
df = df.rename(columns={'index':'tanggal'})
df
```

```python
formatku = {'tanggal':'{:%d/%m/%y}',
            'operasional':'Rp {:.2f}',
            'omset':'Rp {:.2f}'}
laporan = df.style.format(formatku)
laporan
```

```python
type(laporan)
```

```python
laporan.hide_index()
```

```python
laporan.set_caption('Data Omset dan Operasional')
```

```python
laporan.highlight_min('omset', color='pink')
laporan.highlight_max('omset', color='lightgreen')

laporan.highlight_min('operasional', color='lightblue')
laporan.highlight_max('operasional', color='grey')
```

# Merge dua Data Frame secara berdampingan

```python
d1 = {'col1':[1,2,3],
      'col2':[10,20,30]}
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {'col3':[4,5,6],
      'col4':[40,50,60]}
df2 = pd.DataFrame(d2)
df2
```

```python
df = pd.merge(df1, df2, left_index=True, right_index=True)
df
```
